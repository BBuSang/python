```python
import pandas as pd
url = "https://raw.githubusercontent.com/pia222sk20/python_src/refs/heads/main/data/titanic.csv"
df = pd.read_csv(url)
df.rename(columns={'Sex': 'Gender'}, inplace=True)
df
```

```python
df.describe(include='object')
```

```python
# null check
df.isna().mean().sort_values(ascending=False)
```

```python
df.isna().mean().sort_values(ascending=False).plot()
```

```python
temp = df.isna().mean()
temp[temp > 0]
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(4, 2))
temp = df.isna().mean()
temp[temp > 0].sort_values().plot(kind='bar')
plt.show()
```

```python
import missingno as msno

fig, axes = plt.subplots(1,2)
axes[0].set_title('matrix')
axes[1].set_title('bar')
msno.matrix(df=df, figsize=(4,4), color=(0.5,0.5,0.5),fontsize=8 , ax=axes[0])
msno.bar(df=df, figsize=(4,4), color=(0.5,0.5,0.5),fontsize=8 , ax=axes[1])
```

```python
# age컬럼의 결측치가 있는 모든 row를 출력
df[df['Age'].isna()]
```

```python
# 생존유무  pie 그래프
f,ax = plt.subplots(1,2,figsize=(10,5))
ax[0].set_title('생존확율')
df.Survived.value_counts().plot.pie(autopct='%1.1f%%',shadow=True,explode = [0,0.1],ax=ax[0])

ax[1].set_title('생존인원')
df.Survived.value_counts().plot(kind='bar',ax=ax[1])
plt.show()
```

```python
# 선실등급별 생존유무
df.groupby(['Pclass'])['Survived'].sum()
```

```python
pd.crosstab(df.Pclass, df.Survived, margins=True).style.background_gradient(cmap='summer_r')
```

```python
import seaborn as sns
sns.countplot(data=df, x='Pclass', palette=['red','blue','green'],legend=True, hue='Survived')
plt.show()
```

```python
sns.countplot(data=df, x='Gender',hue='Survived', palette=['orange','lightblue'])
plt.show()
```

```python
# 선실등급별, 성별에 따라서 생존확률이 높다는것을 알 수 있음..
# 두 특성을 한꺼번에 비교하는 시각화 툴
# factorplot
sns.catplot(data=df, x='Pclass',y='Survived', hue='Gender', kind='point', palette='muted',aspect=1.5, height=4)
```

```python
# 나이가 가장 낮은 승객, 높은 승객, 평균나이
df.Age.min(), df.Age.max(), df.Age.mean()
```

```python
sns.kdeplot(data=df[df.Survived==1], x='Age',label='Survived==1', fill=True)
sns.kdeplot(data=df[df.Survived==0], x='Age',label='Survived==0', fill=True)
plt.legend(loc='best')
```

```python
# 위의 표를 응용해서 Pclass별로 생존유무에 대한 시각화
sns.catplot(data=df, x='Pclass',y='Survived', hue='Pclass', kind='bar', palette='muted',aspect=1.5, height=4)
```
