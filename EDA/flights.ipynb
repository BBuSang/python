```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
flights = sns.load_dataset('flights')
flights.head()
```

```python
flights.info()
```

```python
flights.T
```

```python
flights.pivot(index='year', columns='month', values='passengers')
```

```python
# 연도별 승객수의 변화 line plot
sns.lineplot(data=flights, x='year', y='passengers', hue="month", errorbar=('ci', 95))
# sns 시각화는 신로구간도 같이 표시하는 경우
# 데이터가 변동될 수 있는 구간
# 신뢰구간이 넓으면 불확실성이 크다... (데이터 분산이 크거나 샘플수가 작을  경우)
# 신뢰구간이 좁으면 안정적... (데이터 모여 있거나 샘플수가 큰 경우)
plt.show()
```

```python
# 연도별 월별 승객의 변화 시각화
plt.figure(figsize=(10,6))
sns.lineplot(data=flights, x='year', y='passengers', hue="month", marker='o')
plt.show()
```
