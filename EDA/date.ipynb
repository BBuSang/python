```python
import pandas as pd
from sklearn.preprocessing import OneHotEncoder
import numpy as np
# 라이브러리 불러오기

# 날짜 형식의 문자열로 구성되는 리스트 정의
ts_ms = pd.date_range(start='2025-01-01', end='2025-10-02', periods=100, tz='Asia/Seoul')
data = {'date': ts_ms.strftime('%Y-%m-%d')}
df = pd.DataFrame(data)
df['date'] = pd.to_datetime(df['date'])
df['year'] = df['date'].dt.year
df['month'] = df['date'].dt.month
df['day_int'] = df['date'].dt.weekday
def convert_week(x) :
    temp = ['월', '화', '수', '목', '금', '토', '일']
    return temp[x]
df['day_str'] = df['day_int'].apply(convert_week)
df['day_weekend'] = df['day_str'].isin(['토', '일']).astype(int)
df
```
